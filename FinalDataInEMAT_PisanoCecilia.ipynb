## Final Data Analysis Report: Comparing Dog and Cat Breeds

The goal of this project was to explore both dog and cat breeds and their differing physicality and behavior by analyzing data from TheDogAPI and TheCatAPI. From the API data, I wanted to understand the common ground between the two species: variation in dog sizes, how cats live longer, and the personality tendencies for each species. These questions shaped the motivation I had and the analytical approach by pushing the project towards comparing the study of weight, lifespan, and temperament across the species and in their breeds. Previous studies and datasets that examine dogs and cats are made separately, but the report I focused on brought both species to light, which motivated this analysis.

The data came from two public API endpoints that compile both breeds of information from reputable registers and veterinary sources. The API returns a list of breed objects, with the objects representing a single breed and containing details on the weight, lifespan and behaviors, and in more cases their origin of history. Although the raw structure is consistent with the data, the fields needed to be restructured into a readable form. For example, weight and lifespan were in textural ranges of ‘7-13’, and temperament was stored in comma-separated strings, meaning the data format needed to be cleaned. In many situations, missing or incomplete data were found in several of the breeds. For the purpose of this report, I focused on key variables that almost every breed had, which ultimately led and addressing the research questions of: weight, lifespan, species, and temperament traits.

Using Python, pandas, and matplotlib, I conducted analyses including boxplots to compare weight and lifespan across species, scatterplots to examine the relationship between weight and lifespan within species, and bar charts to visualize the most common temperament traits. The data is required to be cleaned to ensure consistency throughout the project. Weight and lifespan are examples of stored text ranges, so cleaning is required to get a numeric average for accuracy and better reading. Temperament data needs to be split comma comma-separated strings into individual traits. Cleaned datasets used are dogs_clean and cats_clean; these formed the foundation for all the analyses and visualization used for this project.

Once the data has been cleaned, it can be used for the questions of the analysis. The first question was focused on the average breed weight, which revealed a constant in the two: dog breeds have a very large variety of size range from a few pounds to more than a hundred pounds, while cat breeds fall into a more narrow and consistent band with very small variation. The comparison was visualized in the Preliminary Draft through a bar chart that shows the distribution of average weights across both species, and then was compared as well. The second analysis compared average lifespans, revealing that cats tend to live longer and show more lifespan consistency, while dogs display a wider range with a clear decrease in lifespan among larger breeds. The visualization illustrating this distribution made the contrast between the two even more apparent.

The third analysis I examined was the relationship between weight and lifespan within each species. Scatterplots helped identify the relationship between dogs and cats, showing that with dogs, there was a negative correlation; heavier breeds tended to live shorter lives. While with cats, there is no meaningful relationship between weight and longevity. This difference highlighted the impact of selecting breeding in dogs and especially in the larger working dog group. The final analysis was focused on temperament traits. After extracting the counting temperament words with both breeds in cats and dogs, I created a frequency table that revealed the most common descriptors for each species in highest to lowest format. Dogs tended to be described with socially oriented traits such as “Energetic”, “Friendly”, and “Alert”, while cats had more independent traits of “Affectionate”, “Curious”, and “Intelligent”. A DataFrame table of the top traits made it easy to compare patterns across the two species side-by-side in my previous checkpoints.

Taken together, the results show that dogs exhibit greater variety in size, temperament, and life patterns than cats do. This can be largely related to the highly specialized breeding of dogs in history, which has resulted in distinct categorization made into roles of “Working”, “Toy”, which has made an extreme diversity gap in relation to cats. Cats have undergone less selective diversification, resulting in more stable physical traits and more general behavior descriptors. The findings reinforce the long-time held assumptions about both species, but also reveal the data-driven assumptions, such as the relationship between weight and lifespan in cats, that challenge the common expectations in society.

Of course, the analysis is not without limitations. The breed-level averages the obscure individual variations, lifepsna estimates may not reflect the real-world conditions, and the temperament descriptors are subjective rather than standardized measures. Additionally, the API data can differ in completeness across breeds, meaning some breeds are incomplete, which can provide inaccurate results in some ways. Future work could incorporate the individual animal data and examine the geographic tendencies of breeds on their origins or their historical influences on breed traits, or compare temperament clusters more systematically.

Despite the limitation, this project demonstrates that a comparative, structured analysis can reveal meaningful distinctions between dog and cat breeds by combining visualization, cleaned data, and focused research questions. The report shows clearly how these species differ in size, longevity, and personality characteristics. The main takeaway is that dogs and cats are not just simply different in behavior. They diverge fundamentally in their physical variability, lifespan dynamics, and the ways humans describe and understand them. This analysis offers a foundation for exploration into why these differences exist and how they shape the human-animal relationships.
